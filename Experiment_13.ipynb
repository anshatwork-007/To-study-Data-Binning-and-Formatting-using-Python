```python
## Data Binning--->process of converting continuous numerical data into discrete categories or groups
### ---> helps in simplifying large datasets and makes patterns more visible.
```

```python
import pandas as pd
import numpy as np

Dataset = {
    "Student" : ["A","B","C","D","E","F"],
    "Marks" : [99,88,77,66,87,98],
    "Attendance" : [75,90,23,78,67,100],
    "Grade" : ["A","B","C","D","B","A"]
}
df = pd.DataFrame(Dataset)
print(df)

# Data Binning
print("\nData Binning : ")
Range = [60,70,80,90,100] # Range of marks
Category = ["Below Avg","Average","Above Avg","Top"]
df["Grade"] = pd.cut(df["Marks"],bins=Range,labels=Category)
print(df)

# Binning the attendance of students
print("\nBinning attendance : ")
New_Range = [0,25,50,74,100]
New_Category = ["Too Low","Low","Concerning","Safe"]
df["Attendance"] = (pd.cut(df["Attendance"],bins=New_Range,labels=New_Category))
print(df)
```

```python
# Data formatting:

# Coversion of data type
import pandas as pd
import numpy as np

Dataset = {
    "Student" : ["A","B","C","D","E","F"],
    "Marks" : [99,88,77,66,87,98],
    "Attendance" : [75,90,23,78,67,100],
    "Grade" : ["A","B","C","D","B","A"]
}
df = pd.DataFrame(Dataset)
print(df)

print("\nData Formatting : ")
print(df.dtypes)
df["Attendance"] = df["Attendance"].astype(float) # Converting the attendance's integer value to float values [Int ---> Float]
print(df)

# Rounding Values Of Attendance
df["Attendance"] = df["Attendance"].round(2)
print(df)

# Additional Operations
# Sorting data
sorted_df = df.sort_values(by="Marks")
print(sorted_df)

# Displaying Unique Values
print(df["Attendance"].unique())
```

```python
# Dataset ---> Fruit Dataset
import pandas as pd
import numpy as np
Fruits = {
    "Fruit" : ["Apple","Banana","Mango","Orange","Grapes"],
    "Price" : [150,60,400,200,70],
    "Calories" :[130,500,270,60,420],
    "Weight" : [100,50,200,60,80]
}

df = pd.DataFrame(Fruits)
print(df)

# Binning By Price Of Fruits
print("\nData Binning : ")
Range = [0,100,200,400]
Categories = ["Cheap","General","Expensive"]
df["Price"] = pd.cut(df["Price"],bins=Range,labels=Categories)
print(df)

# Binning By Calories Of Fruits
New_Range = [0,100,200,500]
New_Category = ["Low","Moderate","High"]
df["Calories"] = pd.cut(df["Calories"],bins=New_Range,labels=New_Category)
print("\n",df)
```
